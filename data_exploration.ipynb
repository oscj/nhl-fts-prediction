### Load Oilers Game Stats

```python
oilersGameStats = pd.read_csv("csv/oilersGameStatistics.csv")
```

```python
oilersGameStats["totalGoals"] = oilersGameStats["edmonton_goals"] + oilersGameStats["against_goals"]
```

```python
oilersGameStats = oilersGameStats.drop(columns=['edmonton_goals', 'against_goals'])
```

```python
oilersGameStats.drop(oilersGameStats.tail(1).index,inplace=True)
```
